```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
df=pd.read_csv("airline-passengers.csv",index_col=0,parse_dates=True)
df.head()
```

```python
df.plot()
```

```python
add_decomposition = seasonal_decompose(df, model='addititve')
plt.rcParams['figure.figsize']=[12,8]
add_decomposition.plot()
plt.show()
```

```python
mul_decomposition = seasonal_decompose(df, model='multiplicative')
plt.figure(figsize=(12,8))
mul_decomposition.plot()
plt.show()
```

```python
from random import randrange
from pandas import Series
from matplotlib import pyplot
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
series = [i+randrange(10) for i in range(1,100)]
result = seasonal_decompose(series, model='additive', period=1)
result.plot()
pyplot.show()
```

```python
series = [i**2.0 for i in range(1,100)]
result = seasonal_decompose(series, model='multiplicative', period=1)
result.plot()
pyplot.show()
```
