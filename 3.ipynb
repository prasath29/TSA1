Stationary Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
series=pd.read_csv('daily-total-female-births.csv',header=0,index_col=0)
series.head()
```

```python
series.plot(figsize=(20,10))
plt.figure(figsize=(5,5))
plt.show()
```

```python
series.hist(figsize=(20,10),grid=False)
plt.show()
```

```python
series.mean()
```

```python
X = series.values
split = round(len(X) / 2)
X1, X2 = X[0:split], X[split:]
mean1, mean2 = X1.mean(), X2.mean()
var1, var2 = X1.var(), X2.var()
print('mean1=%f, mean2=%f' % (mean1, mean2))
print('variance1=%f, variance2=%f' % (var1, var2))
```

```python
from statsmodels.tsa.stattools import adfuller
```

```python
result=adfuller(X)
print('ADF Statistics: %f'%result[0])
print('p-value: %f'%result[1])
print('Critical Values: ')
for key,value in result[4].items():
    print('\t%s: %.3f'%(key, value))
```

Non-Stationary Time series Data

```python
series2 = pd.read_csv('airline-passengers.csv', index_col=0)
series2.head()
```

```python
series2.plot(figsize=(20,10))
plt.show()
```

```python
Y=series2.values
split2=round(len(Y)/2)
Y1, Y2 = Y[0:split2], Y[split2:]
Y_mean1, Y_mean2 = Y1.mean(), Y2.mean()
y_var1, y_var2 = Y1.var(), Y2.var()
print('mean1=%f, mean2=%f'%(Y_mean1, Y_mean2))
print('variance1=%f, variance2=%f'%(y_var1,y_var2))
```

```python
y=np.log(Y)
plt.hist(y)
plt.show()
plt.plot(y)
plt.show()
```

```python
result1=adfuller(Y)
print('ADF Statistic: %f' % result1[0])
print('p-value: %f' % result1[1])
print('Critical Values:')
for key, value in result1[4].items():
 print('\t%s: %.3f' % (key, value))
```
